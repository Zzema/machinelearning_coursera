```python
data=pd.read_csv('aucs.txt','\t')
```

```python
import scipy.stats as stats
```

```python
k=0
wil=[]
cols=data.columns[1:]
for i in range(4):
    for j in range(4):
        if i>j: 
            data[k]=data[cols[i]]-data[cols[j]] 
            wil.append(stats.wilcoxon(data[k])[1])
            k=k+1
data.head()
```

```python
data.describe()
```

```python
wil
```

```python
from statsmodels.sandbox.stats.multicomp import multipletests
```

```python
multipletests(wil,alpha = 0.05, method = 'holm') 
```

```python
multipletests(wil,alpha = 0.05, method = 'fdr_bh') 
```
