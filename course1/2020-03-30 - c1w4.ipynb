### 1. Выбрать распределение

Выбрал распределение Лапласа, используется при обработке сигналов

```python
#импорты модулей
import numpy as np
import math
from matplotlib import pylab as plt
```

```python
alpha, beta =2, 1
def func_plotnost (x):
    alpha, beta =2, 1
    return 0.5*alpha*(np.exp(-1*alpha*np.absolute(x-beta)))
def func_rasp (x):
    alpha, beta =2, 1
    return np.where(np.array(x)>2, 1-0.5*np.exp(-1*alpha*(x-beta)), 0.5*np.exp(alpha*(x-beta)))

x1 = np.arange(-1, 3, 4/1000)
y_plotnost=func_plotnost(x1)
y_rasp=func_rasp(x1)
```

```python
plt.plot(x1,y_plotnost,x1,y_rasp)
plt.legend(('plotnost_rasp','func_rasp'))
plt.title('Теоретический вид функции и плотности')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 2. Для выборок из 3 нарисуем функцию распределения средних в выборке и сравним с нормальным распределением на графике (теоретическим)

Математическое ожидание и дисперсия распределения Лапласа следующая:
M=beta=1, d=2/alfa^2=0.5

по ЦПТ M=beta=1, D=d/n - то есть чем больше элементов в выборке, тем меньше дисперсия

### n=3


```python
s=np.zeros(1000)
n=3
for i in range(len(s)):
    s[i] = np.random.laplace(beta, 1/alpha, n).mean() # так как scaleб используемый в функции - это обратная величина от alpha
#s[0] = np.random.laplace(beta, alpha, n).mean()

count, bins, ignored = plt.hist(s, 30, normed=True)
sigma=math. sqrt(0.5/n) #По ЦПТ берем дисперсию нормального распределения Лапласа, которая равна 0,5
#, делим на кол-во элементов в выборке n и извлекаем корень.
mu=1 # так как mu для распределения выборки равен среднему для Лапласа по ЦПТ, а он равен единице.
plt.plot(x1, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x1 - mu)**2 / (2 * sigma**2) ),  linewidth=2, color='r')
plt.legend(('нормальное распределение выборки теоретическое','распределение выборки, n=3'))
plt.title('Сравнение распределения выборок n=3 с нормальным распределением теоретическим по ЦПТ')
plt.xlabel('X')
plt.ylabel('Y')
```
