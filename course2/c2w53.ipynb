```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

data=load_digits()
X=pd.DataFrame(data.data)
y=pd.DataFrame(data.target)
X.head()
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle=False, test_size=0.25)
np.shape(X_train), np.shape(y_train),np.shape(X_test), np.shape(y_test)
```

```python
np.linalg.norm(X_train.values-X_test.values, axis=0)
```

```python
import scipy
```

```python
temp=scipy.spatial.distance.cdist(X_train, X_test, metric='euclidean')
np.shape(temp)
euc_norm=pd.DataFrame(temp)#строка - расстояние между первым трейном и всеми тестами. Столбец - расстояние между тестом и всеми трейнами
euc_norm.head(10)
```

```python
y_pred=y_train.iloc[euc_norm.idxmin(axis=0) ]
y_pred.head()
```

```python
from sklearn.metrics import accuracy_score
sc=accuracy_score(y_test,y_pred)
```

```python
file=open('subc2w531.txt','w+')
file.write(str(1-sc))
file.close()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=1000, random_state=1)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred2=clf.predict(X_test)
```

```python
sc2=accuracy_score(y_test,y_pred2)
sc2
```
